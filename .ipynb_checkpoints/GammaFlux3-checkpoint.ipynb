```python
import pickle
from pylab import *
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import h5py
from mpl_toolkits.mplot3d import Axes3D
```

NOTE: I found that unpickling files created with Python 2 requires the special 'encoding='latin1'' option shown above:

https://stackoverflow.com/questions/28218466/unpickling-a-python-2-object-with-python-3

```python
with open('/data/chocula/villaa/PhotoN_SuperSim/possys/flux_data_100k_v2.pkl','rb') as readFile:
      data=pickle.load(readFile,encoding='latin1')

print(data.keys())
```

```python
#trying to customize here, see:
#https://matplotlib.org/users/customizing.html
#matplotlib.rc('figure', figsize=(3.4, 3.4*(4/6)))
mpl.rcParams['lines.linewidth'] = 2

#ticks
mpl.rcParams['xtick.top'] = False
mpl.rcParams['xtick.bottom'] = True
mpl.rcParams['xtick.minor.visible'] = True
mpl.rcParams['xtick.direction'] = 'in'
mpl.rcParams['xtick.major.size'] = 8
mpl.rcParams['xtick.minor.size'] = 4
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['xtick.minor.width'] = 1
mpl.rcParams['xtick.labelsize'] = 22

mpl.rcParams['ytick.left'] = True
mpl.rcParams['ytick.right'] = False
mpl.rcParams['ytick.minor.visible'] = True
mpl.rcParams['ytick.direction'] = 'in'
mpl.rcParams['ytick.major.size'] = 8
mpl.rcParams['ytick.minor.size'] = 4
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['ytick.minor.width'] = 1
mpl.rcParams['ytick.labelsize'] = 22


#error bars
#mpl.rcParams['errorbar.capsize'] = 3

#axis stuff
mpl.rcParams['axes.labelsize'] = 22

#fonts
# Set the font dictionaries (for plot title and axis titles)
title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'fontname':'Arial', 'size':'32'}
legend_font = {'fontname':'Arial', 'size':'22'}

#fonts global settings
mpl.rc('font',family=legend_font['fontname'])
```

```python
#set up a plot
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(111, projection='3d')
#fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),projection='3d')
#ax1 = axes[0]

#lay out the colors
rgb_r = (255/255.0,0/255.0,0/255.0)
rgb_b = (0/255.0,0/255.0,255/255.0)
rgb_o = (255/255.0,127/255.0,0/255.0)
rgb_b = (31/255.0,120/255.0,180/255.0)
rgb_lr = (251/255.0,154/255.0,153/255.0)
rgb_g = (203/255.0,195/255.0,195/255.0)

ax1.scatter(data['x'][data['inout']==2], data['y'][data['inout']==2], data['z'][data['inout']==2], s=1,c='r', marker='o',label='boundary crossings')


#ax1.set_yscale('linear')
ax1.set_xlim(-0.25, 0.25)
ax1.set_ylim(-0.25,0.25)
ax1.set_zlim(0.7,1.0)
ax1.set_xlabel('sim X pos. [m]',labelpad=20,**axis_font) #labelpad is distance of axis to label
ax1.set_ylabel('sim Y pos. [m]',labelpad=20,**axis_font)
ax1.set_zlabel('sim Z pos. [m]',labelpad=20,**axis_font)
#ax1.grid(True)
#ax1.yaxis.grid(True,which='minor',linestyle='--')

```

```python
for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)
```

```python
plt.tight_layout()
#plt.savefig('figures/pube_replot.eps')
plt.show()
```
