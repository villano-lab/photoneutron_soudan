# Position Sensitivity in the NR Spectra of Photoneutron Analysis

I have condensed the data from `0x0004` version of the simulation to see if it matters what x-y position the detector is at. For example z14 and z5 only differ by x-y position, the vertical (z) distance between them and the source is the same. 

```python
import numpy as np
import pandas as pd
```

```python
#play around with some hits data stored in h5 file
#===============to suppress h5py warning see:
#https://github.com/h5py/h5py/issues/961
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import h5py
warnings.resetwarnings()
fz14_ybe = h5py.File("/data/chocula/villaa/PhotoN_SuperSim/ZipSum/ybe_z14_190320_0x0004_1-3B.h5","r")
fz5_ybe = h5py.File("/data/chocula/villaa/PhotoN_SuperSim/ZipSum/ybe_z5_190321_0x0004_1-3B.h5","r")

#for i in fz14_ybe['simdata/']:
#    print(i)

#for i in fz5_ybe['simdata/']:
#    print(i)

print(list(fz5_ybe['simdata/'].keys()))
```

```python
ybe_z14 = {}
ybe_z14['NRhit'] = np.asarray(fz14_ybe['simdata/NRhit'],dtype=np.float64())
ybe_z14['ERhit'] = np.asarray(fz14_ybe['simdata/ERhit'],dtype=np.float64())
ybe_z14['NRedep'] = np.asarray(fz14_ybe['simdata/NRedep'],dtype=np.float64())

ybe_z5 = {}
ybe_z5['NRhit'] = np.asarray(fz5_ybe['simdata/NRhit'],dtype=np.float64())
ybe_z5['ERhit'] = np.asarray(fz5_ybe['simdata/ERhit'],dtype=np.float64())
ybe_z5['NRedep'] = np.asarray(fz5_ybe['simdata/NRedep'],dtype=np.float64())
```

```python
#cuts
ybe_z14['cNR'] = ybe_z14['ERhit']==0
ybe_z5['cNR'] = ybe_z5['ERhit']==0
```

```python
#do some yield calculations
import lindhard as lind
#import numpy as np

y = lind.getLindhardGe(False)

yv = np.vectorize(y)

#voltage and average energy per pair
V=70 #should be derived from input structure
egam = 3.3 #eV per pair

ybe_z14['NRedep_ee'] = ybe_z14['NRedep']*yv(ybe_z14['NRedep'])
ybe_z14['NRedep_t'] = ybe_z14['NRedep'] + (V/egam)*ybe_z14['NRedep_ee']
ybe_z14['edepNR_t'] = np.sum(ybe_z14['NRedep_t'],1)
ybe_z14['edepNR_ee'] = ybe_z14['edepNR_t']/(1+(V/egam)) #electron-equivalent scale for application of res


ybe_z5['NRedep_ee'] = ybe_z5['NRedep']*yv(ybe_z5['NRedep'])
ybe_z5['NRedep_t'] = ybe_z5['NRedep'] + (V/egam)*ybe_z5['NRedep_ee']
ybe_z5['edepNR_t'] = np.sum(ybe_z5['NRedep_t'],1)
ybe_z5['edepNR_ee'] = ybe_z5['edepNR_t']/(1+(V/egam)) #electron-equivalent scale for application of res
```

```python
#add in the resolution from a model
import pnRes as pnr

r = pnr.getRes('z14') #select the default mode for the resolution model by not specifying second param
rv =  np.vectorize(r)

#apply the res
ybe_z14['edepNR_ee_res'] = ybe_z14['edepNR_ee'] + np.random.normal(0,rv(ybe_z14['edepNR_ee']))
ybe_z14['edepNR_t_res'] = (1+(V/egam))*ybe_z14['edepNR_ee_res']  #convert back to eVt if desired
print(np.shape(ybe_z14['edepNR_t_res']))

ybe_z5['edepNR_ee_res'] = ybe_z5['edepNR_ee'] + np.random.normal(0,rv(ybe_z5['edepNR_ee']))
ybe_z5['edepNR_t_res'] = (1+(V/egam))*ybe_z5['edepNR_ee_res']  #convert back to eVt if desired
print(np.shape(ybe_z5['edepNR_t_res']))
```

```python
#make the histogram

emax = 100000 #keVt scale
#make the new capture histogram
ybe_z14['n_res'],ybe_z14['nx_res'] = np.histogram(ybe_z14['edepNR_t_res'][(ybe_z14['cNR'])]/1000.0,100,range=(0,emax/1000.0)) #do it in keV

ybe_z14['xc_res'] = (ybe_z14['nx_res'][:-1] + ybe_z14['nx_res'][1:]) / 2

ybe_z5['n_res'],ybe_z5['nx_res'] = np.histogram(ybe_z5['edepNR_t_res'][(ybe_z5['cNR'])]/1000.0,100,range=(0,emax/1000.0)) #do it in keV

ybe_z5['xc_res'] = (ybe_z5['nx_res'][:-1] + ybe_z5['nx_res'][1:]) / 2
```
